```python
#adding AS to compare with plots in this data
as_df = pd.read_excel('pnas.1616408114.sd03.xlsx')
df['Accelerated Stability'] = as_df['Slope for Accelerated Stability']
df['Name'] = as_df['Name']
df.head()
```

```python
#comment out if outlier is needed
df = df[df['Name'] != 'rilotumumab']
```

```python
#Scaling the target columns to focus on, and plot all of them in bar plots to see general comparison
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import seaborn as sn

target_columns = ['Accelerated Stability', 'volume', 'hyd_moment', 'pI_seq', 'ens_charge', 'patch_ion', 'patch_neg']
scaler = MinMaxScaler()
scaled_values = scaler.fit_transform(df[target_columns])
scaled_df = pd.DataFrame(scaled_values, columns=target_columns)
df[target_columns] = scaled_df
correlations = scaled_df[target_columns].corr()

for target_feature in target_columns:
    correlations_with_target = correlations[target_feature]
    
    plt.figure(figsize=(10, 6))
    sn.barplot(x=correlations_with_target.index, y=correlations_with_target.values, palette='viridis')
    plt.title(f'Correlation of Features with {target_feature}')
    plt.xlabel('Features')
    plt.ylabel('Correlation Coefficient')
    plt.xticks(rotation=45)
    plt.show()
```
